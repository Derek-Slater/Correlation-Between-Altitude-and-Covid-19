```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
'''
Data 1 is Snippet of data of average country altitudes. Retrieved from wikipedia
'''
data1 = pd.read_csv('countryAltitudes.csv')
data1
```

```python
'''
Data 2 is the WHO data that have the total number of covid cases for each different country
'''
data2 = pd.read_csv('WHO-COVID-19-global-table-data.csv')
data2
```

```python
'''
We are creating an index coloum for easyer data manipulation.
'''
data2['index'] = data2.index
data2
```

```python
'''
We only need the WHO Reguib coloums, which represent the the total number of covid cases for each different country since the outbreak.
'''
filter_data2 = data2[["WHO Region","index"]]
filter_data2
```

```python
'''
We want to merge the WHO Region coloum and Elevation coloum by making sure the country coloum is identical to the index coloum.
'''
plot_data = pd.merge(data1,filter_data2, left_on = "Country" , right_on = "index")
plot_data
```

```python
'''
 we go through in every entry and parse out just the first number, and make the string to a number.
 For example “1,884 m (6,181 ft)[2]” to 1884.0
'''
list=[]
for i in plot_data["Elevation"]:
    list.append(float(i.split()[0].replace(',', '')))
list
```

```python
'''
Adding the "Elevation in M" to our table.
'''
plot_data["Elevation in M"] = list
```

```python
'''
We converte WHO Region coloum to number data type.
'''
convert_dict = {'WHO Region': float }
plot_data1 = plot_data.astype(convert_dict)
```

```python
'''
We created a scatter plot using the imported matlab tools.
'''
x = plot_data1["Elevation in M"]
y = plot_data1["WHO Region"]
plt.scatter(x, y,alpha=0.5)
plt.show()
```

```python
'''
we graphed a regression line for our scatter plot.
'''
sns.regplot(x, y, ci=None)
```
